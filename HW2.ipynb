```python
#Reverse Linked List II

def reverseBetween(self, head, m, n):
        """
        :type head: ListNode
        :type m: int
        :type n: int
        :rtype: ListNode
        """
        
        dummy = ListNode(0)
        dummy.next = head;
        pre = dummy
        for i in range(m-1):
            pre = pre.next
        
        cur = pre.next
        nextNode = cur.next
        for i in range(m,n):
            temp = nextNode.next
            nextNode.next = cur
            cur = nextNode
            nextNode = temp
        
        pre.next.next = nextNode
        pre.next = cur
        
        return dummy.next
```

```python
#Reverse Linked List

def reverseList(self, head):
        """
        :type head: ListNode
        :rtype: ListNode
        """
        if head is None:
            return
        
        p1 = None
        p2 = head
        while p2 is not None:
            p3 = p2.next
            p2.next = p1
            p1 = p2
            p2 = p3
            
        return p1

```

```python
# Linked List Cycle
def hasCycle(self, head):
        """
        :type head: ListNode
        :rtype: bool
        """
        
        cache = set()
        p = head
        while p is not None:
            if p in cache:
                return True
            else:
                cache.add(p)
            p = p.next
        
        return False
```

```python
# Linked List Cycle II

def detectCycle(self, head):
        """
        :type head: ListNode
        :rtype: ListNode
        """
        cache = set()
        temp = head
        while temp is not None:
            if temp in cache:
                return temp
            else:
                cache.add(temp)
                temp = temp.next
        
        return None
```

```python
# Remove Duplicates from Sorted List
def deleteDuplicates(self, head):
        """
        :type head: ListNode
        :rtype: ListNode
        """
        pre = head
        while pre is not None:
            cur = pre.next
            if  pre.next is not None and pre.val == cur.val:
                pre.next = cur.next
            else:
                pre = pre.next
        
        return head
            
```

```python
# Merge Two Sorted Lists

def mergeTwoLists(self, l1, l2):
        """
        :type l1: ListNode
        :type l2: ListNode
        :rtype: ListNode
        """
        
        dummy = ListNode(0)
            
        cur = dummy
        
        p1 = l1
        p2 = l2
        
        while p1 is not None and p2 is not None:
            
            if p1.val <= p2.val:
                cur.next = p1
                p1 = p1.next
            else:
                cur.next = p2
                p2 = p2.next
            cur = cur.next
        
        while p1 is not None:
            cur.next = p1
            p1 = p1.next
            cur = cur.next
        
        while p2 is not None:
            cur.next = p2
            p2 = p2.next
            cur = cur.next
            
        return dummy.next
```
